# Fully-Connected Neural Nets
In the previous homework you implemented a fully-connected two-layer neural network on CIFAR-10. The implementation was simple but not very modular since the loss and gradient were computed in a single monolithic function. This is manageable for a simple two-layer network, but would become impractical as we move to bigger models. Ideally we want to build networks using a more modular design so that we can implement different layer types in isolation and then snap them together into models with different architectures.

In this exercise we will implement fully-connected networks using a more modular approach. For each layer we will implement a `forward` and a `backward` function. The `forward` function will receive inputs, weights, and other parameters and will return both an output and a `cache` object storing data needed for the backward pass, like this:


We've only asked you to implement ReLU, but there are a number of different activation functions that one could use in neural networks, each with its pros and cons. In particular, an issue commonly seen with activation functions is getting zero (or close to zero) gradient flow during backpropagation. Which of the following activation functions have this problem? If you consider these functions in the one dimensional case, what types of input would lead to this behaviour?
1. Sigmoid
2. ReLU
3. Leaky ReLU

## Answer:
Functions that lead to gradient vanishing problem during backpropagation:
  1. Sigmoid: consider when the variables get the extreme values (far from 0), then the gradients of the sigmoid function with respect to these variable are relatively small. And by applying backpropagation through the layers, by mutiply the local gradients with the upstream gradients, the gradients of the lower layers get smaller and smaller, make it very difficult for the low layers to be updated, then the training will make no progess.
  2. ReLu: Consider some how the weights is updated in way that always yeilds the non-positive results for some neurons, then there will be no gradients for these neuron and they will be the 'dead neurons' for the rest of the training. However, by some magical, some of the dead neurons could wake up because the way the weights is updated, but large portion of them will be dead for the rest of the training.

In one dimensional case, the types of input that could lead to gradient vanishing:
1. Sigmoid: Extreme input (far from 0)
2. ReLu: non-negative input
3. Leaky RelU: 0 input

## Inline Question 3:

AdaGrad, like Adam, is a per-parameter optimization method that uses the following update rule:

```
cache += dw**2
w += - learning_rate * dw / (np.sqrt(cache) + eps)
```

John notices that when he was training a network with AdaGrad that the updates became very small, and that his network was learning slowly. Using your knowledge of the AdaGrad update rule, why do you think the updates would become very small? Would Adam have the same issue?


## Answer: 
Because the update rule sum over all the previous squared gradients, then when the training runs for a while, this quantity become bigger and bigger and by divide the current gradient by this amount, it become very close to zero, so the update will be very small, that lead the training to be very slow.

For Adam optimizer, it is not the case. Because for Adam, it has a decay rate to penalty the cache along training, so the cache only takes account for the recent gradients, not all from the begining of the training. In other words, we control this quantity by keeping it not too big during training, make sure the parameters are updated in the right direction to minimize the loss and the amout of gradient updated every time is reasonable.
